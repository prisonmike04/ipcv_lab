```python
new_path = 'scene.jpg'
```

```python
image = cv2.imread(new_path,1)
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
image = cv2.resize(image,(1280,1024))
plt.imshow(image)
```

```python
add_img = cv2.add(img,image)
```

```python
plt.imshow(add_img)
```

```python
fig = plt.figure(figsize=(10,7))
fig.add_subplot(1,3,1)
plt.imshow(img)
plt.axis('off')
plt.title("og_harry")


fig.add_subplot(1,3,2)
plt.imshow(image, cmap = 'gray')
plt.axis('off')
plt.title("new_image")


fig.add_subplot(1,3,3)
plt.imshow(add_img, cmap = 'gray')
plt.axis('off')
plt.title("added_image")
```

```python
sub_img = cv2.subtract(img,image)
```

```python
plt.imshow(sub_img)
```

```python
fig = plt.figure(figsize=(10,7))
fig.add_subplot(1,3,1)
plt.imshow(img)
plt.axis('off')
plt.title("og_harry")


fig.add_subplot(1,3,2)
plt.imshow(image, cmap = 'gray')
plt.axis('off')
plt.title("new_image")


fig.add_subplot(1,3,3)
plt.imshow(sub_img, cmap = 'gray')
plt.axis('off')
plt.title("sub_img")
```

```python
final_and = cv2.bitwise_and(img,image)
final_or = cv2.bitwise_or(img,image)
final_not = cv2.bitwise_not(img)
final_xor = cv2.bitwise_xor(img,image)
```

```python
fig = plt.figure(figsize=(10,7))
fig.add_subplot(2,2,1)
plt.imshow(final_and)
plt.axis('off')
plt.title("and")


fig.add_subplot(2,2,2)
plt.imshow(final_or)
plt.axis('off')
plt.title("or")


fig.add_subplot(2,2,3)
plt.imshow(final_not)
plt.axis('off')
plt.title("not")


fig.add_subplot(2,2,4)
plt.imshow(final_xor)
plt.axis('off')
plt.title("xor")

```
